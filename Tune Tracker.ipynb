#7 - Get all albums from the featured Artists

Now that you have extracted a list of artists from your MongoDB collection in Task #6, your next step is to use the Spotify API to retrieve all albums released by each artist. The goal is to build a comprehensive dataset of albums from the artists featured in your selected genre.

The Spotify API provides an [artist_albums](https://spotipy.readthedocs.io/en/2.22.1/?highlight=artist_albums#spotipy.client.Spotify.artist_albums) endpoint that allows you to query for all albums released by a specific artist. To do this, you will need to use the artist_uri you extracted in Task #6.

Your tasks:


1.   Loop through the `artists_data` object and get the `artist_uri` for each artist. This value will be required for you to call the `artist_albums` endpoint to retrieve the artist's albums.
2.   Use the `spotify.artist_albums()` method to fetch the artist's albums. Store the results in a variable called `albums`.
3.   Add the artist's name to each album returned in the response. This is important for keeping track of which album belongs to which artist.
4.   Append each album to a new list called `artists_albums`.
5.   Spotify API works with something called "pagination". Pagination means that within the string response from the API, there will be another set of results contained in the `next` key. This allows us to create consecutive requests from the same element. Your job is to use the `next` [method](https://spotipy.readthedocs.io/en/2.22.1/?highlight=featured_playlists#spotipy.client.Spotify.next) to get the next albums from a given artist.
6.   Make sure to collect the following fields from each album:
* Artist Name
* Album Name
* Release Date
* Total Tracks
* Album Type
* Spotify URL
* Album Genre(s)
* Available Markets
7.   Continue adding the albums to your artists_albums list until you have exhausted all pages of results.



```python
#START YOUR CODE HERE
artists_albums = []

#Looping through the artists_data list from Task #6
for artist in artists_data:

    artist_uri = artist['artist_uri']

    #Using the Spotify API to fetch all albums from the artist
    results = spotify.artist_albums(artist_uri, album_type='album')

    albums = results['items']

    # Getting artist details to retrieve genre
    artist_details = spotify.artist(artist_uri)
    genres = artist_details['genres']  # This will be a list of genres

    #Appending the albums and add the artist's name to each album
    for album in albums:
        album_info = {
            "artist_name": album['artists'][0]['name'],
            "album_name": album['name'],
            "release_date": album['release_date'],
            "total_tracks": album['total_tracks'],
            "album_type": album['album_type'],
            "spotify_url": album['external_urls']['spotify'],
            "genres": genres,
            "available_markets": album["available_markets"]
        }
        artists_albums.append(album_info)

    #Handling pagination to get the next set of albums
    while results['next']:
        results = spotify.next(results)
        for album in results['items']:
            album_info = {
                "artist_name": album['artists'][0]['name'],
                "album_name": album['name'],
                "release_date": album['release_date'],
                "total_tracks": album['total_tracks'],
                "album_type": album['album_type'],
                "spotify_url": album['external_urls']['spotify'],
                "genres": genres,
                "available_markets": album["available_markets"]
            }
            artists_albums.append(album_info)
#END YOUR CODE HERE
```

```python
# Question 1 - How many albums are stored in the collection?
# Answer:
album_count = albums_collection.count_documents({})
print(f"Number of Albums: {album_count}")
```

```python
# Question 2 - Which artist in the collection has the most albums?
# Answer:
from pymongo import DESCENDING
most_albums_artist = albums_collection.aggregate([

    #Grouping by artists
    {'$group':{'_id':'$artist_name', 'album_count': {'$sum':1}}},

    #Sorting by counting the albums
    {'$sort':{'album_count':DESCENDING}},

    #Getting the top result
    {'$limit':1}
])
for artist in most_albums_artist:
    print(f"Artist with most albums in the collection is {artist['_id']} with {artist['album_count']} albums.")
```

```python
# Question 3 - Which artist has the highest average number of tracks per album?
# Answer:
highest_avg_tracks = albums_collection.aggregate([

    #Grouping by the artist and aggregating the average tracks
    {'$group':{'_id':'$artist_name', 'avg_tracks': {'$avg': '$total_tracks'}}},

    #Sorting by average tracks
    {'$sort': {'avg_tracks': DESCENDING}},

    #Getting the top result
    {'$limit': 1}
])

for artist in highest_avg_tracks:
    print(f"Artist with the highest average number of tracks per album is {artist['_id']} with an average of {artist['avg_tracks']} tracks per album.")
```

```python
# Question 4 - What are the 3 most common genres in the collection?
# Answer:
top_genres = albums_collection.aggregate([

    #As the genres are a list, we need to unwind them
    {"$unwind": "$genres"},

    #Grouping the genres by genre and counting the albums
    {"$group": {"_id": "$genres", "count": {"$sum": 1}}},

    #Sorting them
    {"$sort": {"count": DESCENDING}},

    #Getting the top 3 genres
    {"$limit": 3}
])

# Converting results into a DataFrame
df = pd.DataFrame(top_genres)

# Renaming columns for better readability
df.rename(columns={"_id": "Genre", "count": "Number of Albums"}, inplace=True)

# Displaying the DataFrame
print(df.to_string(index=False))
```

```python
# Question 5 - Who are the top 5 artists with the highest average number of tracks per album in the collection? (*Include the Artist Name*)
# Answer:
highest_avg_tracks = albums_collection.aggregate([

    #Grouping by artist name and aggregating the tracks
    {'$group':{'_id':'$artist_name', 'avg_tracks': {'$avg': '$total_tracks'}}},

    #Soring the average tracks
    {'$sort': {'avg_tracks': DESCENDING}},

    #Getting the top 5 results
    {'$limit': 5}
])

# Converting results into a DataFrame
df = pd.DataFrame(highest_avg_tracks)

# Renaming columns for better readability
df.rename(columns={"_id": "Artist", "avg_tracks": "AVG Tracks per Album"}, inplace=True)

# Displaying the DataFrame
print(df.to_string(index=False))
```

```python
# Question 6 - How many albums in the collection have the genre you selected in Task #5?
# Answer:
#Counting the pop albums
genre_count = albums_collection.count_documents({"genres": "pop"})
print(f"Number of albums with Pop genre: {genre_count}")
```

```python
# Question 7 - What are the top 10 albums with the most tracks? (*Include the Artist Name & the album's genre*)
# Answer:
top_10_albums = albums_collection.aggregate([

    #Soring by total tracks
    {'$sort': {'total_tracks': DESCENDING}},

    #Getting the top 10 artists
    {'$limit': 10},

    #Showing the necessary columns only
    {'$project': { 'artist_name': 1, 'album_name': 1, 'total_tracks': 1, 'genres': 1, '_id': 0}}
])

# Converting results into a DataFrame
df = pd.DataFrame(top_10_albums)

# Renaming columns for better readability
df.rename(columns={"artist_name": "Artist", "album_name": "Album", "total_tracks": "Total Tracks", "genres": "Genre"}, inplace=True)

# Displaying the DataFrame
print(df.to_string(index=False))
```

```python
# Question 8 - Which genre has the highest average number of tracks per album?
# Answer:
highest_avg_tracks = albums_collection.aggregate([

    #Grouping by genres and getting the average tracks
    {'$group': {'_id': '$genres', 'avg_tracks': {'$avg': '$total_tracks'}}},

    #Sorting by the average tracks
    {'$sort': {'avg_tracks': DESCENDING}},

    #Getting the top album
    {'$limit': 1}
])
for genre in highest_avg_tracks:
    print(f"Genre with the highest average number of tracks per album is '{genre['_id'][0]}' with an average of {genre['avg_tracks']} tracks per album.")
```

```python
# Question 9 - Which genre in the collection has the most distinct (unique) artists associated with it?
# Answer:
distict_artists = albums_collection.aggregate([
    # Replacing empty genre lists with "Unknown"
    {'$addFields': {'genres': {'$cond': {'if': {'$eq': ['$genres', []]}, 'then': ['Unknown'], 'else': '$genres'}}}},

    # Unwinding genres to count artists for each genre separately
    {'$unwind': '$genres'},

    # Grouping by genre and count distinct artists
    {'$group': {'_id': '$genres', 'unique_artists': {'$addToSet': '$artist_name'}}},

    # Counting the number of unique artists per genre
    {'$project': {'_id': 1, 'unique_artist_count': {'$size': '$unique_artists'}}},

    # Sorting by the number of unique artists in descending order
    {'$sort': {'unique_artist_count': -1}},

    # Getting the top genre with the most unique artists
    {'$limit': 1}
])

for genre in distict_artists:
    print(f"Genre with the most distinct artists is '{genre['_id']}' with {genre['unique_artist_count']} unique artists.")
```

```python
# Question 10 - How many albums have the word "Deluxe Edition" in their title? (*Include the Artist Name*)
# Answer:
#Counting the number of deluxe editions
deluxe_albums = albums_collection.count_documents({"album_name": {"$regex": "Deluxe Edition", "$options": "i"}})
print(f"Number of albums with 'Deluxe Edition' in their title: {deluxe_albums}")
```

You can change the colors and icon if you like. Learn more about this [here](https://python-visualization.github.io/folium/latest/reference.html#folium.map.Marker)


```python
# Initializing a Folium map centered globally
map_ = folium.Map(location=[20, 0], zoom_start=2)
coordinates_cache = {}

#START YOUR CODE HERE
geolocator = Nominatim(user_agent="MapBox") #Name your geolocator

top_3_albums = albums_collection.find({},{'artist_name': 1, 'album_name': 1, 'genres': 1, 'total_tracks': 1,'available_markets': 1, '_id': 0}).sort("total_tracks", DESCENDING).limit(3)    #Create your query using pymongo here

def get_coordinates(country_code):
    if country_code in coordinates_cache:
        return coordinates_cache[country_code]

    try:
        location = geolocator.geocode(country_code, timeout=10)
        if location:
            coords = (location.latitude, location.longitude)
            coordinates_cache[country_code] = coords
            return coords
        else:
            return None
    except GeocoderTimedOut:
        return None

# Creating a dictionary to store markets and associated album info
market_info = {}

# Looping through each album and collect market information
for album in top_3_albums:

    artist_name = album["artist_name"]
    album_name = album["album_name"]
    total_tracks = album["total_tracks"]
    album_genre = album["genres"]
    available_markets = album["available_markets"]

    for market in available_markets:
        if market not in market_info:
            market_info[market] = []

        market_info[market].append({
            "artist_name": artist_name,
            "album_name": album_name,
            "total_tracks": total_tracks,
            "album_genre": album_genre
        })

# Looping through each market to add markers
for market, albums in market_info.items():
    coords = get_coordinates(market)
    if coords:
        # Preparing popup content
        popup_content = f"<b>Market:</b> {market}<br>"
        for album in albums:
            popup_content += (f"Artist: {album['artist_name']}<br>"
                              f"Album: {album['album_name']}<br>"
                              f"Total Tracks: {album['total_tracks']}<br>"
                              f"Genre: {album['album_genre']}<br><br>")

        # Adding a marker with a popup showing all albums for that market
        folium.Marker(
            location=coords,
            popup=popup_content,
            icon=folium.Icon(color="blue", icon="music")
        ).add_to(map_)

    time.sleep(1)

#END YOUR CODE HERE
map_
```
